# Lab TAT Data

```python
#import the libraries
import pandas as pd
import numpy as np
from scipy import stats
from scipy.stats import norm
```

```python
# Load the dataset
data=pd.read_csv('E:\Assignment\Assignment 3 - Hypothesis testing\LabTAT.csv')
data.head()
```

```python
data.describe()
```

```python
'''
Assume Null hyposthesis as Ho: μ1 = μ2 = μ3 = μ4 (There are no significant differences between the groups' mean Lab values)
Thus Alternate hypothesis as Ha: μ1 ≠ μ2 ≠ μ3 ≠ μ4 (There is a significant difference between the groups' mean Lab values)
Here means of more than two population are different, so we are using one-way anova test.
'''
```

```python
lab1=pd.Series(data.iloc[:,0])
lab1
```

```python
lab2=pd.Series(data.iloc[:,1])
lab2
```

```python
lab3=pd.Series(data.iloc[:,2])
lab3
```

```python
lab4=pd.Series(data.iloc[:,3])
lab4
```

```python
# one-way anova
p_value=stats.f_oneway(lab1,lab2,lab3,lab4)
p_value
```

```python
# compare p_value with α = 0.05 (At 5% significance level)
```

```python
'''
 As (p_value=2.1156) > (α = 0.05); Accept Null Hypothesis i.e. μ1 = μ2 = μ3 = μ4 ( Thus, there is no difference in 
 the average Turn Around Time (TAT) of reports of the laboratories on their preferred list.
 '''
```

# Cutlet Dataset

```python
#import the libraries
import pandas as pd
import numpy as np
from scipy import stats
from scipy.stats import norm
```

```python
# Load the dataset
data=pd.read_csv('E:\Assignment\Assignment 3 - Hypothesis testing\Cutlets.csv')
data.head()
```

```python
data.describe()
```

```python
'''
Assume Null hyposthesis as Ho: μ1 = μ2 (There is no difference in diameters of cutlets between two units)
Thus Alternate hypothesis as Ha: μ1 ≠ μ2 (There is significant difference in diameters of cutlets between two units)
2 Sample 2 Tail test applicable
'''
```

```python
unitA=pd.Series(data.iloc[:,0])
unitA
```

```python
unitB=pd.Series(data.iloc[:,1])
unitB
```

```python
# 2-sample 2-tail ttest:   stats.ttest_ind(array1,array2)     # ind -> independent samples
p_value=stats.ttest_ind(unitA,unitB)
p_value
```

```python
p_value[1]     # 2-tail probability
```

```python
# compare p_value with α = 0.05 (At 5% significance level)
```

```python
'''
 As (p_value=0.4722) > (α = 0.05); Accept Null Hypothesis i.e. μ1 = μ2 ( Thus, there is no difference in diameters 
 of cutlets between two units
 '''
```

# Costomer-Order-Form Dataset

```python
#import the libraries
import pandas as pd
import numpy as np
import scipy
from scipy import stats
from scipy.stats import norm
```

```python
# Load the dataset
data=pd.read_csv('E:\Assignment\Assignment 3 - Hypothesis testing\CostomerOrderForm.csv')
data.head()
```

```python
data.describe()
```

```python
Phillippines_value=data['Phillippines'].value_counts()
Indonesia_value=data['Indonesia'].value_counts()
Malta_value=data['Malta'].value_counts()
India_value=data['India'].value_counts()
print(Phillippines_value)
print(Indonesia_value)
print(Malta_value)
print(India_value)
```

```python
alpha=0.05
errorfree = [271,267,269,280]
defective=[29,33,31,20]
count=[errorfree, defective]
print(count)
```

```python
chiStats = scipy.stats.chi2_contingency(count)
Chi_pValue=chiStats[1]
print("p-value is: "+str(Chi_pValue))
print('Interpret by p-Value')
if chiStats[1] < 0.05:
  print('we reject null hypothesis')
else:
  print('we accept null hypothesis')
```
