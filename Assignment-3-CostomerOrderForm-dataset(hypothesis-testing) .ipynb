```python
alpha=0.05
errorfree = [271,267,269,280]
defective=[29,33,31,20]
count=[errorfree, defective]
print(count)
```

```python
chiStats = scipy.stats.chi2_contingency(count)
Chi_pValue=chiStats[1]
print("p-value is: "+str(Chi_pValue))
print('Interpret by p-Value')
if chiStats[1] < 0.05:
  print('we reject null hypothesis')
else:
  print('we accept null hypothesis')
```
