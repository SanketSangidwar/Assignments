# Salary Data

```python
# Import Libraries
import pandas as pd 
import numpy as np 
import seaborn as sns
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
```

```python
#Importing data
salary_train = pd.read_csv('E:\Assignment\Assignment 17 - Support vector machine\SalaryData_Train(1).csv')
salary_test = pd.read_csv('E:\Assignment\Assignment 17 - Support vector machine\SalaryData_Test(1).csv')
```

```python
salary_train.head()
```

```python
salary_test.head()
```

```python
#Preprocessing
from sklearn import preprocessing
number = preprocessing.LabelEncoder()
```

```python
string_columns=["workclass","education","maritalstatus","occupation","relationship","race","sex","native"]
for i in string_columns:
    salary_train[i] = number.fit_transform(salary_train[i])
    salary_test[i] = number.fit_transform(salary_test[i])
```

```python
colnames = salary_train.columns
colnames
```

```python
len(colnames[0:13])
```

```python
# Splitting data into train and test
train_X = salary_train[colnames[0:13]]
train_y = salary_train[colnames[13]]
test_X  = salary_test[colnames[0:13]]
test_y  = salary_test[colnames[13]]
```

```python
# Kernel = poly
model_poly = SVC(kernel = "poly")
model_poly.fit(train_X,train_y)
pred_test_poly = model_poly.predict(test_X)
```

```python
np.mean(pred_test_poly==test_y) # Accuracy = 78%
```

```python
# kernel = rbf
model_rbf = SVC(kernel = "rbf")
model_rbf.fit(train_X,train_y)
pred_test_rbf = model_rbf.predict(test_X)
```

```python
np.mean(pred_test_rbf==test_y) # Accuracy = 79.64%
```

# Forest Fire

```python
# Import Libraries
import pandas as pd 
import numpy as np 
import seaborn as sns
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
```

```python
# Import Dataset
dataset = pd.read_csv(r'E:\Assignment\Assignment 17 - Support vector machine\forestfires.csv')
dataset.head()
```

```python
dataset.shape
```

```python
dataset.describe()
```

```python
data = dataset.values
X = data[:,2:30]
y = data[:,30]
```

```python
train_X, test_X, train_y, test_y = train_test_split(X,y,test_size=0.3)
```

```python
# kernel = linear
model_linear = SVC(kernel = "linear")
model_linear.fit(train_X,train_y)
pred_test_linear = model_linear.predict(test_X)
```

```python
np.mean(pred_test_linear==test_y)  # Accuracy = 99.35%
```

```python
# Kernel = poly
model_poly = SVC(kernel = "poly")
model_poly.fit(train_X,train_y)
pred_test_poly = model_poly.predict(test_X)
```

```python
np.mean(pred_test_poly==test_y) # Accuracy = 76.28%
```

```python
# kernel = rbf
model_rbf = SVC(kernel = "rbf")
model_rbf.fit(train_X,train_y)
pred_test_rbf = model_rbf.predict(test_X)
```

```python
np.mean(pred_test_rbf==test_y) # Accuracy = 74.35
```
